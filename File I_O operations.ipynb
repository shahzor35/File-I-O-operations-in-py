File operations in py

```python
# Reading a file
file = open("/content/Toyata land cruiser.txt","r")
content = file.read()
print(content)
file.close()
```

```python
file = open("/content/Toyata land cruiser.txt","r")
content = file.read(23)
print(content)
file.close()
```

```python
# Using with statement
# with is used to close file automatically
with open("/content/Toyata land cruiser.txt","r") as file:
  content = file.read()
print(content)
```

```python
# Writing into a file
file_1 = open("/content/Sample.txt","w")
file_1.write("This is first line\n")
file_1.write("This is secound line\n")
file_1.close()
file_1 = open("/content/Sample.txt","r")
content = file_1.read()
file_1.close()
print(content)
```

```python
# Appending data into the existing file
file_2 = open("/content/Sample.txt","a")
file_2.write("This is third line\n")
file_2.write("This is fourth line\n")
file_2.close()
file_2 = open("/content/Sample.txt","r")
content = file_2.read()
file_2.close()
print(content)
```

```python
# Reading JPG file as binary
with open("/content/2018-toyota-land-cruiser.jpg", "rb") as file:
  binary_data = file.read()
with open("/content/2018-toyota-land-cruiser_1.jpg","wb") as file:
  file.write(binary_data)
```

```python
# Checking a file exists or not using os
import os
if os.path.exists("/content/Sample.txt"):
  print("File exists")
else:
  print("File does not exists")
```

```python
# Checking a file exists or not using os
import os
if os.path.exists("/content/Sample2.txt"):
  print("File exists")
else:
  print("File does not exists")
```

```python
# Checking a file exists or not using path library
from pathlib import Path
file = Path("/content/Sample.txt")
if (file.is_file()):
  print("File exists.")
else:
  print("File does not exists")
```

```python
# Checking a file exists or not using path library
from pathlib import Path
file = Path("/content/Sample2.txt")
if (file.is_file()):
  print("File exists.")
else:
  print("File does not exists")
```

```python
# Importing CSV file and printing it line by line
import csv
with open("/content/cars.csv") as csv_file:
  reader = csv.reader(csv_file)
  for row in reader:
    print(row)
```

```python
# Creating a CSV file
import csv
with open("/content/user.csv","w",newline='') as file:
  writer = csv.writer(file)
  writer.writerow(["Name", "Age", "City"])
  writer.writerow(["A", "21", "Hyd"])
  writer.writerow(["B", "22", "Nzb"])

# Printing user.csv file
with open("/content/user.csv") as file:
  content = csv.reader(file)
  for content in content:
    print(content)
```
